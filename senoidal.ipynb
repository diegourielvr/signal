<h5>La frecuencia de una señal de entrada es un problema importante.Es un parámetro muy
relevante en sistemas eléctricos para indicar el balance dinámico entre
generación y consumo de electricidad. En algunas aplicaciones

<h4><strong>¿Qué es una señal senoidal?</strong></h4>
<h5>
<p>
Una señal senoidal, o sinusoide es la gráfica de la función matemática seno de la
trigonometría. Consiste en una frecuencia única con una amplitud constante. <br>En
su forma matemática más simple, una ecuación de Voltaje Senoidal es:
<br><em><strong>V = Vmax Sen q(w*t+φ)</strong></em>
</p>
</h5>
<h4><strong>Parámetros que definen a la señal senoidal</strong></h4>
<p>
<h5>
<ol>
<li>amplitud (o valor de pico)</li>
<li>pulsación (o frecuencia angular)</li>
<li>fase (o ángulo de fase)</li> 
</ol>
El período es el tiempo que tarda la señal en repetirse y su inversa (la
frecuencia) nos indica el número de ciclos que la señal efectúa por unidad de tiempo.
El ángulo de fase nos indica en qué instante de tiempo comienza
la señal a propagarse; al ser una magnitud relativa su significado se entiende mejor cuando
se habla de más de una señal <br>
Otros parámetros importantes son:
<lo>
<li>valor medio</li>
<li>valor eficaz</li>
<li>tiempos de subida y bajada</li>
</p>
<p>
Se denomina frecuencia de la señal a la cantidad de ciclos que pueden desarrollarse en un segundo. Se mide en ciclos por segundo o Hertz, abreviado, Hz. 

La amplitud de la señal se refiere al valor máximo que esta alcanza. Es la distancia máxima entre el punto más alejado de una onda y el punto de equilibrio o medio.

Amplitud de pico:es el valor máximo que tiene una señal, considerada desde el valor "0".

Amplitud pico a pico: es la diferencia entre el valor máximo y el valor mínimo de una señal.

### Instrucciones

En el siguiente código se epueden modificar los parámetros de la función senoidal que se encuentra en la siguiente celda, unicamente se deben modificar los valores que se encuentran a la derecha de los parámetros (amplitud, frecuencia, fase). Y Posteriormente ejecutar el código para ver los cambios reflejados.


```python
from math import pi
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def seno(a: float, f: float, p: float, t: float) -> float:
    return a*sin(2*pi*f*t + p)

# ----- Parámetros -----
amplitud = 1.0
frecuencia = 1.0
fase = 1.0
# ----- fin parámetros -----

x_max = 2*pi
x_min = 0
step = 0.01
x = [i*step+x_min for i in range(0, int((x_max-x_min)//step))]
y = [seno(amplitud, frecuencia, fase, time) for time in x]
fig, ax = plt.subplots()
ax.plot(x,y)
ax.spines['bottom'].set_position('center') # Mover eje inferior al centro
ax.spines[['left', 'right', 'top']].set_visible(False) # Ocultar ejes
ax.set_xmargin(0)
yticks = np.arange((-1*amplitud), amplitud+.5, 0.5)
ax.set_yticks(yticks)
plt.show()
```

<h4><strong>Señales periódicas</strong></h4>
<h5><p>Las señales periódicas son aquellas a las  cuales se les puede encontrar un patrón de  repetitividad, es decir, que después de un determinado tiempo, vuelve a repetirse uno a  uno los valores anteriores, una y otra vez. A este patrón se lo reconoce como ciclo de la onda. El tiempo que demora un ciclo en desarrollarse se denomina período, y por supuesto, se mide en segundos.</p></h5>
<h4><strong>¿Por qué una señal senoidal es periódica?</strong></h4>
<h5><p>Una señal senoidal es periodica si y solo si su frecuencia es un número racional<br>
<strong> Demostración</strong><br>
Por definición x(n) es periódica si y sólo si ∃ N ( N>0 ) tal que:
x(n + N) = x(n) ∀n
<br>El menor valor de N que verifica esta propiedad se denomina
período fundamental.
<br>Para el caso de una onda senoidal, tendríamos:<br>
cos(2πf (N+n) + θ ) = cos(2πf n + θ ) ∀n
<br>que se verifica si y sólo si:<br> 2πf N = 2π k con k entero
<br>= ⇔f=K/N ⇔f es racional<br>
Para determinar el período fundamental N de una senoide, expresamos a la frecuencia como el cociente de dos números
enteros primos relativos. Por lo que el período (N) es el
denominador de esta expresión.</p></h5>

### Instrucciones

En la siguiente celda si módificamos el valor del parámetro periodo se puede apreciar en la gráfica cual es el tiempo que tarda un ciclo completo

```python
from math import sin, pi
import numpy as np
import matplotlib.pyplot as plt

def f_per(a: float, pr: float, p: float, t: float) -> float:
    return a*sin(2*pi*(1/pr)*t + p)

# ----- Parámetros -----
periodo = 2
# ----- fin arámetros -----

amplitud = 1.0
fase = 0

x_max = 4*pi
x_min = -4*pi
step = 0.01
x = [i*step+x_min for i in range(0, int((x_max-x_min)//step))]
y = [f_per(amplitud, periodo, fase, time) for time in x]

fig, ax = plt.subplots()
ax.plot(x,y)
ax.spines[['bottom', 'left']].set_position('center') # Mover eje inferior al centro
ax.spines[['right', 'top']].set_visible(False) # Ocultar ejes
ax.set_xmargin(0)

yticks = np.arange((-4*amplitud), amplitud+4, 2)
ax.set_yticks(yticks)

ax.annotate('periodo = {0}'.format(periodo), 
            xy=(periodo/4, amplitud), xycoords='data',
            xytext=(periodo*(5/4), amplitud), textcoords='data',
            arrowprops=dict(arrowstyle="<->",
                            ec="k",
                            shrinkA=0, shrinkB=0))

plt.show()
```

<h4><strong>¿Qué es frecuencia?</strong></h4>
<h5><p>Es una magnitud que mide el número de repeticiones por unidad de tiempo de cualquier fenómeno, suceso o proceso periodico. La frecuencia se puede medir en diversas unidades. De forma habitual se mide en hercios (Hz) e indica el número de veces que un fenómenos se repite por segundo. También se utilizan la revoluciones por minuto (rpm).<br>Cuenta con número de oscilaciones, que está entendida como un movimiento que se reproduce de un lado a otro naciendo de una posición de equilibrio, es decir, es un ciclo, que es el trayecto que se debe recorrer desde una posición hasta otra para luego regresar, de manera tal que pasa dos veces por la posición de equilibrio. Asimismo, es el número de ciclos por segundo,suele medirse en hercios (Hz).
Las ondas electromagneticas son un tipo de oscilación, pues es una forma de propagación de la radiación electromagnética a través del espacio.
<br> La frecuencia también cuenta con vibraciones, que es el movimiento oscilatorio de un cuerpo provocado por la oscilación de las partículas que lo forman, respecto a una posición de equilibrio.</p></h5>

### Instrucciones

Si modificamos el parametro frecuencia, entonces en gráfica se puede observar que el número de ciclos que hay en un segundo es igual al valor indicado. Unicamente agregar valores enteros positivos

```python
from math import sin, pi
import numpy as np
import matplotlib.pyplot as plt

def seno(a: float, f: float, p: float, t: float) -> float:
    return a*sin(2*pi*f*t + p)

# ----- Parámetros -----
frecuencia = 4
# ----- fin parámetros -----

amplitud = 1.0
fase = 0
periodo = 1/frecuencia

x_max = 1.1
x_min = 0
step = 0.01
x = [i*step+x_min for i in range(0, int((x_max-x_min)//step))]
y = [seno(amplitud, frecuencia, fase, time) for time in x]

fig, ax = plt.subplots()
ax.plot(x,y)
ax.spines[['bottom']].set_position('center') # Mover eje inferior al centro
ax.spines[['left', 'right', 'top']].set_visible(False) # Ocultar ejes
ax.set_xmargin(0)

ax.set_yticks([])
ax.set_xticks([0, 1.0])
inicio = 0
fin = periodo
for n in range(frecuencia+1):
    ax.annotate('', 
            xy=(inicio, amplitud), xycoords='data',
            xytext=(fin, amplitud), textcoords='data',
            arrowprops=dict(arrowstyle="|-|",
                            ec="k",
                            shrinkA=0, shrinkB=0))
    inicio = fin
    fin = (n+1)*periodo

ax.annotate('Frecuencia = {0} Hz'.format(frecuencia), 
            xy=(0, -amplitud), xycoords='data',
            xytext=(frecuencia*periodo, -amplitud), textcoords='data',
            arrowprops=dict(arrowstyle="|-|",
                            ec="k",
                            shrinkA=0, shrinkB=0))
plt.show()
```

<h4><strong>¿Qué es amplitud?</strong></h4>
<h5><p> Es una medida de la variación máxima del desplazamiento u otra magnitud física que varía periódica o cuasiperiódicamente en el tiempo. Es la distancia entre el punto más alejado de una onda y el punto de equilibrio o medio. <br> En otras palabras, es la mitad del valor máximo de pico a pico en una función periódica, como un movimiento armónico simple. Está se mide en metros y es usada para medir las variaciones periódicas que se aprecian a lo largo de su trayectoria, y en las matemáticas, se trata como una medida para observar el valor más alto o bajo de una variable.

Cuando modificamos el parametro amplitud en la siguiente celda, podemos observar en la gráfica que el tamañp máximo que alcanza varia de acuerdo al valor indicado. Estoe es por que la función seno devuelve valores como máximo de 1 y al multiplciarlo por un escalar que en este caso es la amplitud, el tamaño de la onda aumenta o disminuye dependiendo de la amplitud.


```python
from math import sin, pi
import numpy as np
import matplotlib.pyplot as plt

def f_per(a: float, pr: float, p: float, t: float) -> float:
    return a*sin(2*pi*(1/pr)*t + p)

# ----- Parámetros -----
amplitud = 1.0
# ----- fin parámetros -----

periodo = 2*pi
fase = 0

x_max = 2*pi
x_min = -2*pi
step = 0.01
x = [i*step+x_min for i in range(0, int((x_max-x_min)//step))]
y = [f_per(amplitud, periodo, fase, time) for time in x]

fig, ax = plt.subplots()
ax.plot(x,y)
ax.spines[['bottom', 'left']].set_position('center') # Mover eje inferior al centro
ax.spines[['right', 'top']].set_visible(False) # Ocultar ejes
ax.set_xmargin(0)

yticks = np.arange((-amplitud), amplitud+1, 1)
ax.set_yticks(yticks)

ax.annotate('amplitud = {0}'.format(amplitud), 
            xy=(periodo/4, 0), xycoords='data',
            xytext=(0, amplitud), textcoords='data',
            arrowprops=dict(arrowstyle="<->",
                            ec="k",
                            shrinkA=0, shrinkB=0))

plt.show()
```

<h4><strong>¿Qué es fase?</strong></h4>
<h5><p>Es la diferencia de fase entre dos ondas senoidales, usualmente debido a que en el circuito existen capacitores (condensadores) o inductores (bobinas), estos conceptos se refieren a cuando estan conectados entre sí, adquiere carga eléctrica cuando es sometido a un potencial eléctrico con respecto a otro en estado neutro, esto por su propiedad de ser un conductor.<br>
Por lo tanto, la fase o el  ángulo de fase nos permite indicar cuanto se adelanta o se atrasa una onda con respecto a otra.
</p></h5>

### Instrucciones

Si el parámetro fase de la señal senoidal se módifica, entonces se podrá observar en la gráfica que ocurre un desplzamiento de la señal con respecto a la señal sonidal con fase = 0

```python
#Grafica de periodo
from math import sin, pi
import numpy as np
import matplotlib.pyplot as plt

def seno(a: float, f: float, p: float, t: float) -> float:
    return a*sin(2*pi*f*t + p)

# ----- Parámetros -----
fase = 2
# ----- fin parámetros -----

amplitud = 1.0
frecuencia = .15
periodo = 1/frecuencia

x_max = 2*pi
x_min = -2*pi
step = 0.01
x = [i*step+x_min for i in range(0, int((x_max-x_min)//step))]
y = [seno(amplitud, frecuencia, 0, time) for time in x]
y2 = [seno(amplitud, frecuencia, fase, time) for time in x]

fig, ax = plt.subplots()
ax.plot(x,y)
ax.plot(x,y2)
ax.spines[['bottom', 'left']].set_position('center') # Mover eje inferior al centro
ax.spines[['right', 'top']].set_visible(False) # Ocultar ejes
ax.set_xmargin(0)

yticks = np.arange((-amplitud), amplitud+1, 1)
ax.set_yticks(yticks)

ax.annotate('fase = {0}'.format(fase), 
            xy=((periodo/4)-fase, amplitud), xycoords='data',
            xytext=((periodo/ 4), amplitud), textcoords='data',
            arrowprops=dict(arrowstyle="|-|",
                            ec="k",
                            shrinkA=0, shrinkB=0))

plt.show()
```

<h4><strong>Longitud de onda</strong></h4>
<h5><p>Recorre una perturbación periódica que se propaga por un medio en un ciclo. La longitud de onda, también conocida como periodo espacial es la inversa de la frecuencia multiplicado por la velocidad de propagación de la onda en el medio por el cual se propaga, es decir, describe lo larga que es la onda.La longitud de onda es una distancia real recorrida por la onda. La letra griega λ (lambda) se utiliza para representar la longitud de onda en ecuaciones.
<h4><strong>Relación con la frecuencia</strong></h4>
Si la velocidad de propagación es constante, la longitud de onda λ es inversamente proporcional a la frecuencia f. Una longitud de onda más larga corresponde a una frecuencia más baja, mientras que una longitud de onda más corta corresponde a una frecuencia más alta:
<br> λ= v/f<br>
donde λ es la longitud de onda, v es su velocidad de propagación, y f es la frecuencia.
<br>El espectro electromagnético es una distribución en función del rango de frecuencias de las ondas o radiación electromagnética.
Este espectro se ha dividido convencionalmente en bandas o segmentos definidos a partir de unos límites de frecuencia o longitud de onda.
<br>Dentro de un medio homogéneo, cualquier onda viaja a velocidad constante. Sin embargo, cuando se produce un cambio en el medio la onda experimenta un cambio en su velocidad y también un su dirección debido a la refracción.
La relación entre la longitud de onda al pasar de un medio A a otro medio B puede expresarse en función de los índices de refracción de los dos medios.
</p></h5>

### Instrucciones

En la siguiente celda cuando modificamos el parámetro de longitud de onda, se puede apreciar la distancia que hay entre dos valores máximos de la señal senoidal.

```python
from math import sin, pi
import numpy as np
import matplotlib.pyplot as plt

def f_per(a: float, pr: float, p: float, t: float) -> float:
    return a*sin(2*pi*(1/pr)*t + p)

# ----- Parámetros -----
longitud_de_onda = 4
# ----- fin parámetros -----
amplitud = 1.0
fase = 0

x_max = 2*pi
x_min = -2*pi
step = 0.01
x = [i*step+x_min for i in range(0, int((x_max-x_min)//step))]
y = [f_per(amplitud, longitud_de_onda, fase, time) for time in x]

fig, ax = plt.subplots()
ax.plot(x,y)
ax.spines[['bottom', 'left']].set_position('center') # Mover eje inferior al centro
ax.spines[['right', 'top']].set_visible(False) # Ocultar ejes
ax.set_xmargin(0)

yticks = np.arange((-2*amplitud), amplitud+2, 2)
ax.set_yticks(yticks)

ax.annotate('longitud de onda = {0}'.format(longitud_de_onda), 
            xy=(longitud_de_onda/4, amplitud), xycoords='data',
            xytext=(longitud_de_onda*(5/4), amplitud), textcoords='data',
            arrowprops=dict(arrowstyle="<->",
                            ec="k",
                            shrinkA=0, shrinkB=0))

plt.show()
```

<h4><strong>Señal compuesta</strong></h4>
Una señal compuesta es una única unidad, un grafo de dominio de frecuencia muestra la señal compuesta  como  una serie de frecuencias compuestas.
Una señal analógica compuesta está formada por múltiples ondas seno.

<strong>Espectro de frecuencia</strong>
<br>Es la combinación de todas las ondas seno, que componen esa señal.  
<br>En la descomposición generada de señales, la señal de frecuencia más baja se denomina frecuencia fundamental o primer armónico. 
<br>El resto de armónicos serán múltiplos enteros de la frecuencia fundamental. 
<br>Si la señal es aperiódica, la descomposición da una combinación de ondas seno con frecuencias continuas.

<h5>La importancia del estudio de las señales se debe a que los circuitos son alimentados por fuentes que poseen duiferentes formas de onda y estas representan las difefrentes señales que poseen dversas formas de onda.


<h4><strong>Importancia de las señales senoidales y compuestas</strong></h4>
<h5><p>Los generadores industriales son de fuerza electromotriz sinusoidal, por lo que, la red de suministri de energía eléctrica transporta corrientes que varían con el tiempo según esta función.
<dr>El estudio de la respuesta de los circuitos sinusoidal nos da la respuesta a otro tipo de sañales, pues cualquier señal se puede desarrollar en suma de componentes sinusoidales, y por lo tanto la suma de las repsuestas individuales nos da la respuesta global.
Una señal senoidal es una corriente o tensión que varía sinusoidalmente con el tiempo.
<br>Mientras que la importancia de las señales compuestas es que representa como se comporta una onda, por ejemplo: Las ondas sonoras se crean por vibraciones y el sonido se puede grabar en discos de vinilo con surcos que hacen que la aguja de un tocadiscos vibre. Luego, esas vibraciones se amplifican para reproducir el sonido que se grabó originalmente. El tono, o la frecuencia, de ese sonido puede adquirir una amplia gama de valores. En otras palabras, la frecuencia de ese sonido es una variable continua. La señal producida al reproducir un disco se puede representar como una onda de sonido suave que describe exactamente cómo cambia el sonido con el tiempo. Esto se conoce como señal analógica. La onda de sonido producida por el tocadiscos es análoga o similar a la onda de sonido original que se registró, esto hablando de una señal analógica compuesta.

</p></h5>

<h4><strong>Ancho de banda</strong></h4>
<h5><p>El ancho de banda se refiere al rango de las frecuencias componentes y el espectro de frecuencia está Dominios del tiempo y de la frecuencia relacionado con los elementos dentro de ese rango, es decir, es la diferencia entre la frecuencia más alta y más baja contenidas en la señal.
<br>
Una onda seno de frecuencia única no es útil para transmitir datos.  
Es necesario usar una señal compuesta, una señal formada por múltiples ondas seno (señales periódicas simples). 
<br>De acuerdo con el análisis de Fourier, cualquier señal compuesta es realmente una combinación de ondas simples con distintas frecuencias, amplitudes y fases. </p></h5>
<h4><strong>Transformada de Fourier</h4></strong>
<h5><p>La transformada de Fourier es una operación matemática fundamental para algunas disciplinas como las telecomunicaciones o la física. Sin ella no existirían las telecomunicaciones modernas, no solo Internet o la telefonía móvil, sino la propia telefonía convencional, que no habría podido evolucionar más allá de una forma de comunicación local y no habrían existido las llamadas de larga distancia. 
<br>Sin entrar a valorar las matemáticas detrás de ellas, lo que permite la transformada de Fourier es convertir cualquier función matemática a otro domino, denominado el dominio de la frecuencia. Esto facilita tratar y analizar las funciones de una forma alternativa. Por ejemplo: <br>Un ingeniero de sonido puede tener una grabación de audio con mucho ruido, por ejemplo porque había un pitido en el ambiente. Estos ruidos muchas veces pueden eliminarse, pero para ello habría que separar la parte del sonido deseada y la que se quiere eliminar. Esta separación muchas veces es imposible por métodos convencionales, ya que la señal que se tiene está ya sumada, y sin saber una de las partes es imposible eliminar la otra.
<br>
<h4><strong>Espectro de frecuencia</strong></h4>
Un tono fundamental que se combine con varios armónicos de distintos niveles se percibe como un sonido. Las relaciones de nivel entre estos elementos sónicos cambian con el tiempo. La combinación de un número de armónicos se conoce como el espectro armónico, más comúnmente conocido como el espectro de frecuencias.
<br>El espectro de frecuencias muestra todos los elementos sónicos individuales de un sonido. Se muestra de bajo a alto y de izquierda a derecha a lo largo del tiempo. Los respectivos niveles de todos los armónicos se reflejan verticalmente, y los picos más altos indican los niveles más elevados.</p></h5>
